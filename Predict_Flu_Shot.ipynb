```python
# Read data
data_temp = pd.read_csv('driven_data/training_set_features.csv')

data_temp_test = pd.read_csv('driven_data/test_set_features.csv')
```

```python
data_temp_label = pd.read_csv('driven_data/training_set_labels.csv')
```

```python
train_objs_num = len(data_temp)
dataset = pd.concat(objs=[data_temp, data_temp_test], axis=0)
dataset_preprocessed = pd.get_dummies(dataset)
data_temp = dataset_preprocessed[:train_objs_num]
data_temp_test = dataset_preprocessed[train_objs_num:]
```

```python
train_data = data_temp.values
train = data_temp_label
```

```python
# Method 1: One Vs Rest
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
```

```python
model = Pipeline([
                ('clf', OneVsRestClassifier(LogisticRegression(solver='sag'), n_jobs=-1)),
            ])
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

v = ['h1n1_vaccine', 'seasonal_vaccine']

acc = []
s_fold = StratifiedKFold(n_splits=5)

from sklearn.preprocessing import Normalizer

scaler = Normalizer().fit(train_data)
train_data = scaler.transform(train_data)
```

```python
print(model)
for cl in v:
    train_label = np.array(train[cl])
    for train_index, test_index in s_fold.split(train_data, train_label):
        traindata, validdata = train_data[train_index], train_data[test_index]
        trainlabel, validlabel = train_label[train_index], train_label[test_index]
        model.fit(traindata, trainlabel)
        y_pred = model.predict(validdata)
        acc.append(accuracy_score(validlabel, y_pred))
    print("Average")
    print(np.mean(acc)*100.0)
    print("\n")
```
